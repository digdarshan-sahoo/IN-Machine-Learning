## **Question 1.1**

```python
import math
from abc import ABCMeta, abstractmethod

class AbstactBaseShape(metaclass=ABCMeta):
    def __init__(self): pass
    @abstractmethod
    def accept_attributes(self): pass
    @abstractmethod
    def calc_area(self): pass
    @abstractmethod
    def display_area(self): pass
    
class Triangle(AbstactBaseShape):
    def __init__(self):
        self._side1 = ""
        self._side2 = ""
        self._side3 = ""
        self._area = ""
        
    def accept_attributes(self):
        print("Enter the length of 3 sides of a TRIANGLE:")
        self._side1 = float(input("Enter side1 : "))
        self._side2 = float(input("Enter side2 : "))
        self._side3 = float(input("Enter side3 : "))

    def calc_area(self):
        print("Calculating area of the triangle...")
        tri_peri = self._side1 + self._side2 + self._side3
        tri_halfperi = tri_peri / 2
        self._area = math.pow(tri_halfperi * 
                    (tri_halfperi-self._side1) * 
                    (tri_halfperi-self._side2) * 
                    (tri_halfperi-self._side3),
                    0.5)
         
    def display_area(self):
        print("The area of the triangle is {:.4f} units.".format(self._area))
    

try:
    TriangleObj = Triangle()
    TriangleObj.accept_attributes()
        
except Exception as e:
    print("{0}:Exception: Object initialisation error: {1}".format(type(e),e))
else:
    print("\nObject initialised successfully...\n")
    try:
        TriangleObj.calc_area()
        TriangleObj.display_area()
              
    except Exception as e:
        print("{0}:Exception: Error with area: {1}".format(type(e),e))
finally:
    del TriangleObj
    print("\nObject deleted successfully...")      
```

## Question 1.2 

```python
from abc import ABCMeta, abstractmethod

class AbstactBaseLong(metaclass=ABCMeta):
    def __init__(self): pass
    @abstractmethod
    def filter_long_words(self): pass
    @abstractmethod
    def display(self): pass
    
class FilterLongWords(AbstactBaseLong):
    def __init__(self):
        self._filtered_animals = []
        self._len = ""
        
    def filter_long_words(self, animals, word_len):
        self._filtered_animals = [a for a in animals if len(a)>word_len]
        self._len = word_len
        self.display()

    def display(self):
        print("The words bigger than {} letters are:\n {}\n".format(self._len, self._filtered_animals))
    
animals = ["Ox",
           "Fox","Cow",
           "Deer","Lion",
           "Tiger","Sheep",
           "Donkey","Jaguar",
           "Giraffe","Dolphin",
           "Elephant","Anaconda",
           "Rhinoceros","Wildebeest",
           "Albatross", "Alligator",
           "Caterpillar","Grasshopper"]
                         
try:
    WordObj = FilterLongWords()
        
except Exception as e:
    print("{0}:Exception: Object initialisation error: {1}".format(type(e),e))
else:
    print("\nObject initialised successfully...\n")
    try:
        print("Entire List: \n {} \n".format(animals))
        WordObj.filter_long_words(animals,4)
        WordObj.filter_long_words(animals,7)
        WordObj.filter_long_words(animals,10)
              
    except Exception as e:
        print("{0}:Exception: Error filtering long words: {1}".format(type(e),e))
finally:
    del WordObj
    print("\nObject deleted successfully...")      
```

```python
from abc import ABCMeta, abstractmethod

class AbstactBaseLength(metaclass=ABCMeta):
    def __init__(self): pass
    @abstractmethod
    def word_length_map(self): pass
    @abstractmethod
    def display(self): pass
    
class ListWordsLength(AbstactBaseLength):
    def __init__(self):
        self._wordlen_l = []
        
    def word_length_map(self, animal_list):
        self._wordlen_l = [len(a) for a in animal_list]
        self.display()

    def display(self):
        print("The words length in the list is: \n {} \n".format(self._wordlen_l))
    
animals = ["Ox",
           "Fox","Cow",
           "Deer","Lion",
           "Tiger","Sheep",
           "Donkey","Jaguar",
           "Giraffe","Dolphin",
           "Elephant","Anaconda",
           "Albatross", "Alligator",
           "Rhinoceros","Wildebeest",
           "Caterpillar","Grasshopper"]
                         
try:
    WordObj = ListWordsLength()
        
except Exception as e:
    print("{0}:Exception: Object initialisation error: {1}".format(type(e),e))
else:
    print("\nObject initialised successfully...\n")
    try:
        print("Input Words List: \n {} \n".format(animals))
        WordObj.word_length_map(animals)
              
    except Exception as e:
        print("{0}:Exception: List mapping error: {1}".format(type(e),e))
finally:
    del WordObj
    print("\nObject deleted successfully...")      
```

```python
from abc import ABCMeta, abstractmethod

class AbstactBaseAlpha(metaclass=ABCMeta):
    def __init__(self): pass
    @abstractmethod
    def is_vowel(self): pass
    @abstractmethod
    def check_alpha(self): pass
    
class CheckAlphabet(AbstactBaseAlpha):
    def __init__(self):
        self._alpha = ""
        
    def check_alpha(self, alpha):
        if not alpha.isalpha():
            raise TypeError(" You must enter an alphabet.")
        if len(alpha) != 1:
            raise ValueError(" Enter a value with length 1, passed({}).".format(len(alpha)))
        

    def is_vowel(self, alpha):
        self.check_alpha(alpha)
        vowel = {'a','e','i','o','u','A','E','I','O','U'}
        if alpha in vowel:
            return True
        else:
            return False
    
                         
try:
    letter = input("Enter an alphabet : ")
    AlphaObj = CheckAlphabet()
        
except Exception as e:
    print("{0}:Exception: Object initialisation error: {1}".format(type(e),e))
else:
    print("\nObject initialised successfully...\n")
    try:
        check_vowel = AlphaObj.is_vowel(letter)
        if check_vowel is True:
            print("\n You entered a Vowel\n")
        else:
            print("\n You entered a Consonant\n")
            
              
    except Exception as e:
        print("{0}:Exception: Vowel checking error: {1}".format(type(e),e))
finally:
    del AlphaObj
    print("\nObject deleted successfully...")      
```
