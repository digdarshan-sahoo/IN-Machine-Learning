## Question 1:
1. Write a program which will find all the numbers which are divisible by 7 but are not multiple of 5, between 2000 and 32000 (both included).
The numbers obtained should be printed in a comma-separated sequence on a single line.

```python
from abc import ABCMeta, abstractmethod
```

```python
class AbstactNumSeriesBase(metaclass=ABCMeta):
    def __init__(self): pass
    @abstractmethod
    def find_num_series(self): pass
    @abstractmethod
    def check_range(self): pass
    
class NumSeries(AbstactNumSeriesBase):
    def __init__(self,low,high):
        self._min_range = int(low)
        self._max_range = int(high)
        self._range_OK = self.check_range()
        #print("self._range_OK:", self._range_OK)
        
    def check_range(self):
        if self._min_range >= self._max_range:
            print("Invalid Range...")
            print("Range must start with lower value(given{%d}) and end with higher value (given{%d})" \
                  % (self._min_range, self._max_range))
            return False
        else:
            return True
    
    def find_num_series(self):
        if self._range_OK == True:
            for num in range(self._min_range,self._max_range+1,1):
                if (num%7==0) & (num%5!=0):
                    print('{}, '.format(num),end='')

try:
    NumObj = NumSeries(2000,3200)
except TypeError as te:
    print("TypeError: Pass a range of two values:",te)
except ValueError as ve:
    print("ValueError: Pass two integer values only:",ve)
except Exception as e:
    print("Exception: Object initialisation error",e)
else:
    print("Object initialised successfully...")
    try:
        NumObj.find_num_series()
    except Exception as e:
        print("Exception: Series generation error:",e)
finally:
    print("\nNo execution beyond this...")
```

```python

from abc import ABCMeta, abstractmethod
class AbstactNamesBase(metaclass=ABCMeta):
    def __init__(self): pass
    @abstractmethod
    def check_valid_names(self): pass
    @abstractmethod
    def reverse_names(self): pass
    @abstractmethod
    def print_names(self): pass
    
class ReverseName(AbstactNamesBase):
   
    def __init__(self):
        self.get_names()
        #print("self._first_name:",self._fname,"self._last_name:", self._lname)
        self._valid_names = self.check_valid_names()
        #print("self._valid_names:", self._valid_names)
    
    def get_names(self):
        self._fname = input("Enter your first name: ")
        self._lname = input("Enter your last name:")
    
    def check_valid_names(self):
        if self._fname.isalpha() == False:
            print("Invalid First Name... ")
            print("First name{%s} must contain only alphabets." \
                  % (self._fname))
            return False
        elif self._lname.isalpha() == False:
            print("Invalid Last Name... ")
            print("Last name{%s} must contain only alphabets." \
                  % (self._lname))
            return False
        else:
            return True
    
    def reverse_names(self):
        if self._valid_names == False:
            raise ValueError("Either of the names are not alphabets.")
        if self._valid_names == True:
            self._fname = self._fname[::-1]
            self._lname = self._lname[::-1]
        print("Names reversed successfully...")
    def print_names(self):
        print("Your name in reverse order is: {} {} \n".format(self._lname, self._fname))    
        

try:
    RevNameObj = ReverseName()
except AttributeError as ae:
    print("AttributeError: Incorrect attribute is in use:",ae)
except TypeError as te:
    print("TypeError: Datatype of Name is not string or wrong formatting option passed:",te)
except ValueError as ve:
    print("ValueError: A name is not alphabetic:",ve)
except Exception as e:
    print("AdditionalError: During object creation:",e)

else:
    print("Object initilised succefully...\n")
    try:
        RevNameObj.reverse_names()
    except ValueError as ve:
        print("ValueError: Error while reversing:", ve)
    except AttributeError as ae:
        print("AttributeError: Error while reversing:", ae)
    except Exception as e:
        print("AdditionalError: During reversing the names:",e)
    else:
        RevNameObj.print_names()

finally:
    print(RevNameObj.__dict__)
```

```python
from abc import ABCMeta, abstractmethod
import math

class AbstractShape(metaclass=ABCMeta):
    def __init__(self): pass
    @abstractmethod
    def calc_volume(self): pass
    @abstractmethod
    def display(self): pass
    
class Sphere(AbstractShape):
   
    def __init__(self,d):
        self._diameter = float(d)
        self._pie = math.pi
        
    def calc_volume(self):
        self._radius = self._diameter / 2
        self._volume = (4.0/3.0) * self._pie * (self._radius * self._radius * self._radius)
        
    
    def display(self):
        print("The volume of the Sphere with diamter{%.2f}cm = {%.2f} cubic cm. \n" \
                  % (self._diameter, self._volume))
        

try:
    s = Sphere(12)
except AttributeError as ae:
    print("AttributeError: Incorrect attribute is in use:",ae)
except TypeError as te:
    print("TypeError: Datatype of diameter is not desired datatype:",te)
except ValueError as ve:
    print("ValueError: Diameter does not hold proper value:",ve)
except Exception as e:
    print("AdditionalError: During object creation:",e)

else:
    print("Object initilised succefully...\n")
    try:
        s.calc_volume()
    except ValueError as ve:
        print("ValueError: Error while calculating volume:", ve)
    except AttributeError as ae:
        print("AttributeError: Error while calculating volume:", ae)
    except Exception as e:
        print("AdditionalError: While calculating volume:",e)
    else:
        s.display()

finally:
    print(s.__dict__)

    
```
