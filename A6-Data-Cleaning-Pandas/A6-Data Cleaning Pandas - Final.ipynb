## Solve Using OOPs Concept.

```python
import math
import pandas as pd
import numpy as np

from abc import ABCMeta, abstractmethod

class AbstactBaseFlightData(metaclass=ABCMeta):
    def __init__(self): pass
    @abstractmethod
    def fix_flightnumber_column(self): pass
    @abstractmethod
    def fix_from_to_column(self): pass
    @abstractmethod
    def fix_recentdelays_column(self): pass
    @abstractmethod
    def get_flight_data(self): pass
    
class FlightData(AbstactBaseFlightData):
    def __init__(self, flight_data):
        self._flight_data = flight_data
        #print(self._flight_data)
    def fix_flightnumber_column(self):
        '''
        Business Rules:
        1.a) Some values in the the FlightNumber column are missing. 
        These numbers are meant to increase by 10 with each row so 10055 and 10075 need to be put in place.
        1.b) Fill in these missing numbers and make the column an integer column (instead of a float column).
        '''
        print("Fixing FlightNumber Column...")
        
        temp_flight_no = self._flight_data.loc[:,'FlightNumber']
        
        for (i_index,i_value) in zip(temp_flight_no.index,temp_flight_no):
    
            if math.isnan(i_value):
                temp_flight_no.iloc[i_index] = i_prev_value + 10
            else:
                pass

            i_prev_value = i_value

        self._flight_data['FlightNumber'] = self._flight_data['FlightNumber'].astype(int)
        print("Fixed FlightNumber Column...\n")
        
    def fix_from_to_column(self):
        '''
        Business Rule 2:
        2. The From_To column would be better as two separate columns! Split each string on the 
        underscore delimiter _ to give a new temporary DataFrame with the correct values. 
        Assign the correct column names to this temporary DataFrame.
        '''
        print("Fixing From_To Column...")
        temp_flt_data = self._flight_data.copy()
        from_to_col = temp_flt_data['From_To'].str.split("_",n = 1, expand = True)
        temp_flt_data['Origin'] = from_to_col[0]
        temp_flt_data['Destination'] = from_to_col[1]
        
        '''
        Business Rule 3:
        3. Notice how the capitalisation of the city names is all mixed up in this temporary DataFrame. 
        Standardise the strings so that only the first letter is uppercase 
        (e.g. "londON" should become "London".)
        '''
        temp_flt_data['Origin'] = temp_flt_data['Origin'].str.capitalize()
        temp_flt_data['Destination'] = temp_flt_data['Destination'].str.capitalize()
          
        '''Business Rule 4:
        4. Delete the From_To column from df and attach the temporary DataFrame from the previous questions.
        '''
        del temp_flt_data['From_To']
        self._flight_data = temp_flt_data.copy()
        print("Fixed From_To Column...\n")
        
        
    def fix_recentdelays_column(self):
        ''' Business Rule 5:
        5. In the RecentDelays column, the values have been entered into the DataFrame as a list. 
        We would like each first value in its own column, each second value in its own column, and so on.
        If there isn't an Nth value, the value should be NaN. Expand the Series of lists into a DataFrame 
        named delays, rename the columns delay_1, delay_2, etc. and replace the unwanted 
        RecentDelays column in df with delays.
        '''
        print("Fixing RecentDelays Column...")
        recent_delays_df = pd.DataFrame(self._flight_data['RecentDelays'])
        recent_delays_flat = pd.DataFrame(recent_delays_df['RecentDelays'].values.tolist())
        # Find number of columns in the flattened dataframe. Loop and add the columns to original dataframe.
        for i in range(len(recent_delays_flat.columns)):
            new_col_name = 'delay_' + str(i+1)
            self._flight_data[new_col_name] = recent_delays_flat[0]
        del self._flight_data['RecentDelays']
        print("Fixed RecentDelays Column...\n")
        
    def get_flight_data(self):
        return self._flight_data
    

# Load data to dataframe
flight_data = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm',
                                    'Budapest_PaRis', 'Brussels_londOn'],
                        'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
                        'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
                        'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
                        '12. Air France', '"Swiss Air"']
                        })


try:
    FlightDataObj = FlightData(flight_data)
        
except Exception as e:
    print("{0}:Exception: Object initialisation error: {1}".format(type(e),e))
else:
    print("\nObject initialised successfully...\n")
    try:
        # STEP1: Fix FlightNumber Column
        FlightDataObj.fix_flightnumber_column()
                
        # STEP2: Split and fix From_To Column
        FlightDataObj.fix_from_to_column()
        
        # STEP3: Split RecentDelays column.
        FlightDataObj.fix_recentdelays_column()
        flt_data = FlightDataObj.get_flight_data()
        print("Updated Flight Data:\n",flt_data) 
        
    except Exception as e:
        print("{0}:Exception: Printing Flight Data: {1}".format(type(e),e))
finally:
    del FlightDataObj
    print("\nObject deleted successfully...")      
```
