## **Question (1)**

```python
# Define a user defined function for division
def my_division(x,y):
    if x/y == 0:
        raise ZeroDivisionError
    else:
        print("{}/{} = {:.2f}".format(x,y,x/y))

try:
    print("These divisions will work fine...\n")
    
    my_division(10,2) # Works fine. No error.
    my_division(5,3) # Works fine. No error.
    
    print("\nThis division will throw Exception...\n")
    
    my_division(5,0) # Throws exception

except ZeroDivisionError as de:
    print("{}:RunTime Exception: Ooops.. denominator is zero.: {}".format(type(de),de))
except Exception as ex:
    print("{}:RunTime Exception: Some error occurred:{}".format(type(ex),ex))
else:
    print("No Error occurred during calculations.")
```

## Question (2) 

```python
from abc import ABCMeta, abstractmethod

class AbstactBaseSports(metaclass=ABCMeta):
    def __init__(self): pass
    @abstractmethod
    def display(self): pass
    
class SportsByCountry(AbstactBaseSports):
    def __init__(self, sub, ver, obj):
        self._subjects = sub
        self._verbs = ver
        self._objects = obj
        
    def display(self):
        sentences = ["{} {} {}.".format(s, v, o) \
                     for s in self._subjects for v in self._verbs for o in self._objects]
        for line in sentences:
            print(line)

            
subjects=["Americans","Indians"]
verbs=["play","watch"]
objects=["Baseball","Cricket"]
                         
try:
    SportsObj = SportsByCountry(subjects, verbs, objects)
        
except Exception as e:
    print("{0}:Exception: Object initialisation error: {1}".format(type(e),e))
else:
    print("\nObject initialised successfully...\n")
    try:
        print("Countries that love sports:\n")
        SportsObj.display()
        
    except Exception as e:
        print("{0}:Exception: Error displaying countries that love sports: {1}".format(type(e),e))

finally:
    del SportsObj
    print("\nObject deleted successfully...")      
```
